```python
# MergingJoiningConcatation
```

```python
import numpy as np
import pandas as pd
```

```python
emp = pd.DataFrame({
    'employee id':[1,2,3,4,5],
    'name' :['john','ana','smith','jack','william'],
    'location':['AP','TS','HP',"J & k",'Goa']
})
jobs =pd.DataFrame({
    'employee id':[1,2,3,6,7],
    'Salary':[60000,79900,73927,739278,4897],
    'tax':[20,10,3,6,9]
})
```

```python
emp
```

```python
jobs
```

```python
pd.merge(emp,jobs)
# alteast on column should be match 
```

```python
pd.merge(emp,jobs,on ='employee id')
```

```python
pd.merge(emp,jobs,on ='employee id',how='inner')
```

```python
pd.merge(emp,jobs,on ='employee id',how='outer')
```

```python
pd.merge(emp,jobs,on ='employee id',how='left')
```

```python
pd.merge(emp,jobs,on ='employee id',how='right')
```

```python
dt1=pd.DataFrame({
    'A':['A0','A1','A0'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})
dt2 =pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
    
})
```

```python
dt1
```

```python
dt2
```

```python
pd.concat([dt1,dt2])
```

```python
pd.concat([dt1,dt2],axis=1)
```
