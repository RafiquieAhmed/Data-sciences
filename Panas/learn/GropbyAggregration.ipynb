```python
import pandas as pd

# Create data
data = {
    "Category": ["Electronics", "Grocery", "Clothing", "Electronics", "Grocery"],
    "Store": ["Store A", "Store B", "Store C", "Store A", "Store B"],
    "Sales": [50000, 15000, 20000, 30000, 18000],
    "Quantity": [5, 30, 20, 3, 25],
    "Date": ["2026-01-01", "2026-01-02", "2026-01-03", "2026-01-04", "2026-01-05"],
    "Price": [10000, 500, 1000, 10000, 600],
    "Profit": [8000, 3000, 5000, 6000, 3500]
}

# Create DataFrame
df = pd.DataFrame(data)

print(df)
```

```python
cat=df.groupby('Category')
cat
for  i,v in cat:
    print(i)
    print(v)
```

```python
cat=df.groupby('Category')['Sales'].sum()
cat
```

```python
cat=df.groupby('Store')['Sales'].sum()
cat
```

```python
#create by muliple colums
cat=df.groupby(['Category','Store'])['Sales'].sum()
cat
```
